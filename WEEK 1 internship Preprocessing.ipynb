```python
import math
# Selecting numeric columns
numeric_columns = data.select_dtypes(include=['float64', 'int64'])
# for subplots
num_cols = len(numeric_columns.columns)
num_rows = math.ceil(num_cols / 3) 
# figure and axis
fig, axes = plt.subplots(num_rows, 3, figsize=(15, num_rows * 5))
# Flatten the axes if there's only one row
if num_rows == 1:
    axes = axes.reshape(1, -1)
# Loop through each numeric column and create box plots
for i, col in enumerate(numeric_columns.columns):
    row_index = i // 3
    col_index = i % 3
    sns.boxplot(x=data[col], orient='v', ax=axes[row_index, col_index], color='skyblue')
    axes[row_index, col_index].set_title(f'Box plot of {col}')
    axes[row_index, col_index].set_ylabel(col)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
#handling outliers
from scipy.stats.mstats import winsorize
alpha = 0.05
# Winsorize each numeric column
for col in numeric_columns.columns:
    data[col] = winsorize(data[col], limits=(alpha, alpha))
```

This code will winsorize each numeric column in the DataFrame, replacing extreme values with less extreme values determined by the specified limits.

```python
plt.figure(figsize=(30,25))

# Plot histograms for each numeric column before winsorization
for i, col in enumerate(numeric_columns.columns):
    plt.subplot(2, len(numeric_columns.columns), i + 1)
    sns.histplot(data[col], kde=True, color='skyblue')
    plt.title(f'Before Winsorization: {col}')
    plt.xlabel('')
    plt.ylabel('Frequency')

# Plot histograms for each numeric column after winsorization
for i, col in enumerate(numeric_columns.columns):
    plt.subplot(2, len(numeric_columns.columns), len(numeric_columns.columns) + i + 1)
    sns.histplot(data[col], kde=True, color='salmon')  # Plotting original data for comparison
    plt.title(f'Original Data: {col}')
    plt.xlabel('')
    plt.ylabel('Frequency')


plt.show()
```

```python
import numpy as np
from scipy.stats.mstats import winsorize

# Winsorize numeric columns in the original data
numeric_columns = data.select_dtypes(include=['float64', 'int64'])
winsorized_data = data.copy()  # Create a copy of the original data to preserve it

for col in numeric_columns.columns:
    winsorized_data[col] = winsorize(data[col], limits=[0.05, 0.05])
```

```python
# Numeric columns after winsorization
numeric_columns_winsorized = winsorized_data.select_dtypes(include=['float64', 'int64'])
# for subplots
num_cols = len(numeric_columns_winsorized.columns)
num_rows = math.ceil(num_cols / 3) 
# figure and axis
fig, axes = plt.subplots(num_rows, 3, figsize=(15, num_rows * 5))
# Flatten the axes if there's only one row
if num_rows == 1:
    axes = axes.reshape(1, -1)
# Loop through each numeric column and create box plots
for i, col in enumerate(numeric_columns_winsorized.columns):
    row_index = i // 3
    col_index = i % 3
    sns.boxplot(x=winsorized_data[col], orient='v', ax=axes[row_index, col_index], color='skyblue')
    axes[row_index, col_index].set_title(f'Box plot of {col} (After Winsorization)')
    axes[row_index, col_index].set_ylabel(col)
plt.tight_layout()
plt.show()
```
